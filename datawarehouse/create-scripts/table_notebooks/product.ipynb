```python
# Load db
go_sales = load_db('../source/go_sales.sqlite')

# Load tables
product = load_table(go_sales, 'PRODUCT').add_prefix('PRODUCT_')
product_line = load_table(go_sales, 'PRODUCT_LINE').add_prefix('PRODUCT_LINE_')
product_type = load_table(go_sales, 'PRODUCT_TYPE').add_prefix('PRODUCT_TYPE_')
```

```python
# Merge tables
# Product.product_type_code      = Product_Type.product_type_code
# Product_line.product_line_code = Product_type.product_line_code

result = pd.merge(product, product_type, left_on='PRODUCT_PRODUCT_TYPE_CODE', right_on='PRODUCT_TYPE_PRODUCT_TYPE_CODE')
result = pd.merge(result, product_line, left_on='PRODUCT_TYPE_PRODUCT_LINE_CODE', right_on='PRODUCT_LINE_PRODUCT_LINE_CODE')
result = result.drop(columns=['PRODUCT_PRODUCT_TYPE_CODE', 'PRODUCT_TYPE_PRODUCT_LINE_CODE'])
result
```
